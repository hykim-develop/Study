### 연습

2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Oerder Priority' 컬럼을 제거
3. 컬럼의 이름을 변경 -> 대륙, 국가, 상품종류, 판매채널, 주문일자, 발송일자, 판매개수, 단가, 원가
4. 대륙, 국가를 기준으로 오름차순 정렬
5. 인덱스를 초기화하고 기존의 인덱스는 제거 
6. 파생변수 생성 
    - 총단가 = 단가 * 판매개수
    - 총원가 = 원가 * 판매개수
    - 총이윤 = 총단가 - 총원가

```python
# 원본의 데이터의 복사본을 생성 (copy() 함수를 호출하여 복사본을 생성 )
df = sales.copy()
# 컬럼을 제거 
df.drop(
    ['Order ID', 'Total Revenue', 'Total Cost', 
     'Total Profit', 'Order Priority'], 
     axis = 1, inplace = True
)
```

```python
# drop() 함수를 이용하여 컬럼을 제거하 데이터프레임을 새로운 변수에 대입
# ( 컬럼을 제거한다 + 복사본 데이터를 생성 )
df2 = sales.drop(
    ['Order ID', 'Total Revenue', 'Total Cost', 
     'Total Profit', 'Order Priority'], 
     axis = 1
)
```

```python
# 컬럼의 이름을 변경 -> 2개의 방법을 사용 
# ( rename() 함수 이용 / DataFrame.columns 변수의 데이터를 변경 )
# rename() -> 특정 컬럼의 이름을 변경할 때 주로 사용 ( 컬럼의 이름이 변경된 데이터를 먼저 확인하고 바꾸기 위해 )
# DataFrame.columns의 변수 데이터를 변경할 때는 확인 후 변경이 불가능 ( 모든 컬럼의 이름을 변경 )
df.rename(
    columns = {
        'Region' : '대륙',
        'Country' : '국가', 
        'Item Type' : '상품종류', 
        'Sales Channel' : '판매채널', 
        'Order Date' : '주문일자', 
        'Units Sold' : '판매개수', 
        'Ship Date' : '발송일자', 
        'Unit Price' : '단가', 
        'Unit Cost' : '원가'
    }, inplace = True
)
```

```python
df2.columns = ['대륙', '국가', '상품종류', '판매채널', '주문일자', 
               '발송일자', '판매개수', '단가', '원가']
```

```python
# 대륙, 국가를 기준으로 오름차순 정렬 
# value를 기준으로 정렬을 변경 -> sort_values()
# 잘못된 예시 -> 
    # 1. 대륙을 기준으로 오름차순을 정렬
    # 2. 국가를 기준으로 오름차순을 정렬 -> 1번 작업과 상관 없이 국가별로 오름차순
df.sort_values('대륙', inplace=True)
df.sort_values('국가', inplace=True)
```

```python
df
```

```python
# 대륙별로 오름차순 정렬이 되고 대륙별 국가로 오름차순 정렬이 되는 방법 
# sort_values 함수에 기준이 2개
df.sort_values(['대륙', '국가'], inplace=True)
```

```python
# 인덱스를 초기화하고 기존의 인덱스는 제거 
# reset_index() -> 기존의 인덱스를 제거한다 drop 매개변수를 사용
df.reset_index(drop=True , inplace=True)
```

```python
# 파생변수를 생성 (총단가, 총원가, 총이윤)
# 총단가 , 총원가 -> 기존의 데이터에서 계산식 
# 총이윤 -> 파생변수에서 파생된 새로운 변수 
df['총단가'] = df['단가'] * df['판매개수']      
# 행렬 간의 곱 -> Series, DataFrame이든 values의 타입은 행렬 -> value * value
# -> 행렬 * 행렬 (두개의 행렬의 크기가 같아야 한다.)
df['총원가'] = df['원가'] * df['판매개수']
df['총이윤'] = df['총단가'] - df['총원가']
# df['총이윤'] = ( df['단가'] * df['판매개수'] ) - ( df['원가'] * df['판매개수'] )
```

```python
df.head()
```

```python
# 어떤 대륙이 총이윤의 합계가 가장 클까?
# 대륙별 고유값 리스트를 생성 -> unique() // value_counts()
# 대륙 컬럼을 기준으로 비교연산자를 사용해서 필터링 -> 총이윤의 합산 (반복 실행)
df['대륙'].unique()
```

```python
df['대륙'].value_counts()  # 결과의 타입은? -> Series -> index의 값들은 추출하면 고유값 리스트
```

```python
list_1 = df['대륙'].value_counts().index
```

```python
data_list = []
for data in list_1:
    # data -> 대입이 되는 데이터는? -> 각 대륙의 이름 
    # print(data)
    # 조건식 생성 
    flag = df['대륙'] == data
    # 조건식을 이용하여 필터 -> 총이윤의 합산
    result = df.loc[flag, '총이윤'].sum()
    # print(result)
    # 대륙의 이름과 총이윤의 합산을 dict 형태로 생성하여 새로운 리스트에 추가 
    dict_data = { '대륙' : data, '총이윤 합산': result }
    data_list.append(dict_data)
data_list
```

```python
# data_list를 데이터프레임으로 변경 
pd.DataFrame(data_list)
```

```python
# 데이터의 그룹화 -> 같은 데이터들을 결합하여 연산한 뒤 새로운 데이터로 구성 
# 그룹화 함수 -> groupby( 그룹화를 할 컬럼의 이름(2개 이상이라면 []로 입력) )

group_data = df.groupby('대륙')

group_data
```
