### 시계열 데이터
- 시간을 나타내는 데이터 타입
    - 시간이라는 데이터는 굉장히 복잡한 구조
    - ( 연도(4자리 연도수, 2자리 연도수), 월(숫자형 표시, 문자형 표시), 일, 시(12시 기준/24시 기준), 분, 초, 요일(숫자형 표시, 문자형 표시), 오전/오후 )
    - 위의 시간의 종류가 하나의 데이터 타입에 묶여 있는 데이터 타입
- 시간이라는 데이터의 타입은 python에서 기본 제공이 아니라 라이브러리를 통해서 시계열 데이터를 생성
    - datetime 라이브러리 안에 시계열 데이터 타입이 존재 (해당 라이브러리 따로 설치하실 필요는 없다.)
    - pandas 라이브러리에서도 시계열데이터가 존재함 
- 일반적인 데이터 파일을 로드하면 시계열 데이터는 문자로 취급
    - 문자로 되어있는 데이터 타입을 시계열 데이터로 변환
        - pasing(파싱) -> 내가 사용하기 편하게 데이터 타입은 변환
            - csv 파일을 데이터프레임으로 변환
            - list 형태의 데이터 데이터프레임으로도 변환을 했었다.
            - 문자형 데이터를 시계열 데이터로 변환
            - 웹문서를 soup 형태로 변환
1. 시계열 데이터는 시간 계산이 가능
2. 차순 정렬
3. 데이터 형태로 시계열이라는 하나의 통일된 형태로 구성(데이터 정형화)
4. 시계열데이터에서 문자형 데이터로 변환 (파싱) 과정에서 시계열 데이터의 일부를 추출하여 사용 가능
    - 특정 날짜의 요일을 추출
    - 시간만 따로 추출한다거나
    - 오전 / 오후만 따로 추출한다거나...
    
    - 시계열 데이터의 format(type)
        - 4자리 년도수 : %Y
        - 2자리 년도수 : %y
        - 월을 숫자로 표시 : %m
        - 월을 문자로 표시 : %B
        - 월을 문자로 축약해서 표시 : %b
        - 일을 숫자로 표시 : %d
        - 24시 기준 시를 표시 : %H
        - 12시 기준 시를 표시 : %I
        - AM/PM 표시 : %p
        - 분을 표시 : %M
        - 초를 표시 : %S
        - 요일을 이름으로 표시 : %A
        - 요일을 숫자로 표시 : %w (0: 일, 1:월, ... 6: 토)

```python
# datetime 라이브러리 호출 ( 2가지 호출방식: import datetime // from datetime import datetime ★ 추천 )
from datetime import datetime
# datetime 라이브러리 안에 datetime 기능만 가져오겠다.
```

```python
time_str="2026-04-15"
```

```python
type(time_str)
```

```python
# 문자열데이터를 시계열 데이터로 변경
# datetime 안에 함수를 호출 : 
    # strptime() : 문자열 데이터를 시계열 데이터로 변환
    # 필수 인자 2개
        # 문자열 데이터
        # 해당 문자열의 포멧이 어떻게 되어있는가? (설명)
time_data=datetime.strptime(time_str,"%Y-%m-%d")
```

```python
time_data
```

```python
# 시계열 데이터에서 필요한 부분만 추출 -> 시계열 데이터를 문자 데이터로 변환
# 요일을 추출 : strftime() -> 시계열에서 추출할 데이터의 포멧
time_data.strftime('%y/%m/%d')
```

```python
time_str2='2023년 10월 01일 17시 30분 20초'
# time_str2를 시계열 데이터로 변환
time_data1=datetime.strptime(time_str2,'%Y년 %m월 %d일 %H시 %M분 %S초')
```

```python
# 해당하는 날짜의 요일이 언제인가?
time_data1.strftime("%a")
```

```python
# pandas에서 시계열데이터를 이용하는 방법
import pandas as pd
```

```python
# 시계열 데이터를 생성
# Timestamp() -> 시계열 생성 class
# 1970-01-01을 기준으로 int형 데이터가 들어오면 나노초만큼 증가
pd.Timestamp(1000000)
```

```python
pd.Timestamp(100,unit='D')
```

```python
pd.Timestamp('2026-04-15')
```

```python
pd.Timestamp("10001010091020")
```

```python
# pandas, datetime의 시계열 데이터의 차이는 자동화
# 일반적인 날짜의 포멧을 pandas는 저장을 해놨다. (한글형 포멧은 없다.)
# to_함수() -> 데이터의 타입을 변경하는 함수 to_타입() 해당 타입으로 변경하겠다.
pd.to_datetime('2026-04-15')
```

```python
pd.to_datetime('04/10/26')
```

```python
pd.to_datetime('2026년4월15일',format="%Y년%m월%d일")
```

```python
# 오늘로부터 15일 뒤
# 현재 날짜에서 15일 뒤는 언제인가?
now_date=pd.to_datetime('2026-04-15')
```

```python
# 15일 데이터 생성
pd.Timedelta(15,unit='D')
```

```python
now_date+pd.Timedelta(15,unit='D')
```

```python
pd.Timedelta(hours=10)
```

```python
# 시계열 데이터 - 시계열 데이터?
pd.to_datetime('2026-04-15')-pd.to_datetime('2026-01-01')
```
