### 자료형 데이터
- 여러개의 데이터를 하나의 변수에 대입하기 위한 형태

- tuple
    - 자료형 데이터 중에 가장 기본적인 형태 ( 가장 메모리를 덜 먹는 데이터의 형태 )
    - ()를 이용해서 데이터를 묶여서 표현
    - 한번 생성이 되면 수정이 불가능 ( 원소를 수정하거나 추가 삭제 작업을 할때는 새로운 tuple를 생성 )
- list
    - []를 이용해서 데이터를 묶여서 표현(생성)
    - tuple과는 다르게 각 원소들을 수정하거나 추가, 삭제가 가능함. (리스트에서 제공하는 내장 함수를 이용)
    - 원소의 개수를 0개부터 생성이 가능 ( 비어있는 리스트를 생성 -> 여러 개의 데이터를 하나의 리스트에 추가하여 새로운 데이터의 집합 생성 )
- dic
    - {}를 이용하여 데이터를 묶여서 표현(생성)
    - key : value의 형태로 데이터가 구성 (list에서 index를 key로 표현한 방식)
    - key를 기준으로 데이터를 선택, 수정, 삭제 등이 가능
    - 원소의 개수를 0개부터 생성이 가능

```python
# tuple 형태
# 원소의 타입이 다르더라도 생성이 가능
tuple_a = (10,'hello')
tuple_b = (3.14, True)
print(tuple_a)
print(tuple_b)
print(type(tuple_a))
```

```python
# data -> 4가지 기능 - CRUD 
# Create (생성), Read (조회), Update (갱신), Delete (삭제)
# tuple에서 각각의 원소에 접근 (조회)
# tuple은 CR 중 R에 해당됨
# tuple의 원소들은 index(위치)를 기준으로 추출
# index는 0부터 시작
print(tuple_a[0]) # tuple_a에서 첫번째 원소를 추출하여 출력 (10)
print(tuple_b[1]) # tuple_b에서 두번째 원소를 추출하여 출력 (True)
```

```python
# tuple 끼리의 합 (문자열 끼리의 합과 같이 연결)
tuple_a + tuple_b 
```

```python
# tuple에 숫자를 곱한다면?
tuple_a * 3 
```

```python
# tuple과 숫자의 합은? ( + 기호는 같은 타입의 데이터와 결합 )
tuple_b+3
```

```python
# 특정 영역의 데이터를 추출
tuple_c=(10,20,30,40)
# tuple[ 시작index : 종료index ] -> 시작 index부터 종료 index 전까지의 데이터를 추출
print(tuple_c[1:3]) # 두번째 원소부터 4번째 원소 전까지
```

```python
# 시작 index나 종료 index의 값을 적지 않는다면 처음부터, 마지막까지
print(tuple_c[1:]) # 두번째 원소부터 마지막 원소까지
print(tuple_c[:3]) # 처음부터 네번째 원소 전까지
```

```python
# 마지막에서 2번째부터 마지막까지
print(tuple_c[2:])
print(tuple_c[-2:])
# len() : 길이를 출력
print(tuple_c[len(tuple_c)-2:])
```

```python
# list 데이터 생성 (Create)
list_a = [30,20,40,10]
list_b= ['abc','ad','fi','be']
```

```python
# 에러 발생
# list_a -> list 형태의 데이터
# . -> ~안에 
# sort() -> 리스트 데이터를 오름차순 정렬을 하는 함수 (리스트 안에 함수)
# reverse() -> 리스트 데이터를 역순으로 정렬을 변경하는 함수 (리스트 안에 함수)
# list_a.sort() -> 결과 
list_a.sort().reverse()
```

```python
# sort() 정렬은 변경 // sorted()
sorted(list_a).reverse()
```

```python
# sort() 정렬은 변경 // sorted()
# sorted() 함수는 결과를 리스트로 되돌려주기 때문에 .reverse() 사용이 가능
# 해당 데이터는 저장x -> sorted() 함수의 결과가 출력(저장x) -> reverse() 역순으로 변경은 하는데 어디에 저장할지 모른다. (저장 위치가 없다.)
sorted(list_a).reverse()
```

```python
# list_b는 문자데이터로 이루어진 리스트 -> 오름차순 정렬??
list_b.sort()
print(list_b)
```

```python
# 내림차순 정렬 -> sort() -> reverse()
# sort() -> 오름차순(기본) / 내림차순 둘다 가능 (옵션을 변경)
# sort() 함수 안에 reverse 매개변수의 값을 True 변경하면 내림차순
list_b.sort(reverse=True) 
print(list_b)
```

```python
# 리스트에 데이터를 추가
# append() : 리스트 가장 뒤에 데이터를 대입 -> 데이터
# insert() : 특정 위치에 데이터를 대입 -> 특정 위치, 데이터 
list_a.append(55)
print(list_a)
```

```python
# 특정 위치에 데이터는 대입 -> 대입( 추가, 제거라는 의미는 전혀 존재x )
# 데이터를 다루는 언어들에서는 제거라는 확실한 명령이 없으면 데이터는 제거x
list_a.insert(2,35)
print(list_a)
```

```python
# 리스트에서 원소를 제거
# pop() : 리스트에서 가장 마지막 원소를 제거
list_a.pop()
print(list_a)
```

```python
# list에서 3번째 데이터를 제거하고 싶다 -> 근데 del 키워드가 기억이 나지 않는다.
# list에서 해당 데이터를 확인(value 확인) -> remove() 제거 -> 위험은 존재(같은 value가 앞에 존재하는 경우)
# del을 쓰는게 더 나음 remove는 위험성이 존재해서 

new_list=[]

for i in range(len(list_a)):
    # list_a의 길이만큼 반복 실행
    # i : 리스트의 인덱스 값들
    if i == 1:
        # list_a[1] 항목은 new_list에 추가가 되지 않는다.
        continue
    new_list.append(list_a[i])

print(new_list)
```
