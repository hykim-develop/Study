```python
# break문
# 반복문을 강제적으로 종료
for i in range(10):
    #10번 반복을 하는 반복문 (i는 0부터 대입돼서 9까지)
    #i가 3일 때(조건문) 반복문을 종료
    if i==3:
        break
    print(i)
print('-'*16)
print(i)
```

```python
#while+break
#1부터 1씩 증가시켜서 누적합이 1500000이 넘어가는 시점은 언제이며 누적합은 몇인가?

#누적합->값들이 더해질 공간이 필요
result=0
#while문에서 시작값을 지정
i=1

#while 조건식(bool) -> 반복 횟수가 명확하지 않을때 -> 무한루프 사용 (while True) -> 조건이 충족되면 반복문을 강제종료

while True:
    # result에 i의 값을 더한다.
    # #1 (반복 횟수:1733) - 오답
    # result가 150만보다 크거나 같은가? -> result와 i를 더한다 -> i의 값을 1 증가시킨다
    # 마지막의 전 반복 : result = 1499046 -> 조건식이 거짓 -> result=1499046+1732 -> i = 1732+1 -> 반복문으로 돌아간다.
    # 마지막 반복 :  result = 1500778 -> 조건문이 참 -> 반복문이 종료 (i의 값은 1733)
    # result에 i인 1733은 더해지지 않았다.
    # if result>=1500000:
    #     break
    # result = result + i
    result += i
    # #2 (반복횟수 : 1732회)
    # 마지막 반복 : result = 1499046 -> result = 1499046 + 1732 -> 조건문 참 -> 반복문이 강제 종료
    # if result >= 1500000:
    #     break
    # 시작값을 변화 ( 1씩 증가 )
    i += 1
    # #3
    # if result >= 1500000:
    #    break
#i의 값을 출력 ( 1부터 i까지 반복했다는 출력 값 )
print(i)
# result 값을 출력 (1부터 i까지의 누적합)
print(result)

# while Skip
```

```python
# 2개의 주사위가 존재
# 2개의 주사위의 합이 5의 배수 (5,10)인 모든 경우를 출력하시오.
# (1,4) (2,3) (3,2) (4,1) -> 합이 5인 경우
# (4,6) (5,5) (6,3) -> 합이 10인 경우
# 총 경우의 수는? -> 36개

# 첫번째 주사위의 경우의 수 반복문으로 생성
for i in range(1,7):

    # 두번째 주사위의 경우의 수 반복문으로 생성
    for j in range(1,7):

        # 두 개의 주사위의 합이 5의 배수라면? (조건식 생성)
        k=i+j
        l=k%5 #( (i+j)%5 == 0 )
        if l==0:
            # 조건식이 참인 경우 두 개의 주사위의 숫자를 출력
            print(i,j)
```
