#### 데이터 결합 연습 
1. transaction 데이터프레임과 csv 폴더 안에 item_master 파일을 로드하여 2개의 데이터프레임을 결합 
2. 결합된 데이터에 csv 폴더에 customer_master 파일을 로드 하여 데이터프레임을 결합 
3. payment_date 컬럼의 데이터를 시계열데이터로 변환 ( to_datetime(스리즈) )

```python
# 파일 로드 
item_master = pd.read_csv("../csv/item_master.csv")
customer_master = pd.read_csv("../csv/customer_master.csv")
```

```python
item_master.head(1)
```

```python
customer_master.head(1)
```

```python
item_master.info()
```

```python
customer_master.info()
```

```python
transaction.info()
```

```python
# transaction , item_master  조인 결합 -> 공통 컬럼은 item_id
df = transaction.merge(
    item_master, 
    on = 'item_id', 
    how = 'inner'
)
```

```python
df.info()
# item_master에서 결합된 컬럼 item_name, item_price의 결측치가 존재하지 않는다
# 같은 집합 : item_id 기준으로 inner 결합 -> 인덱스도 유지 / 결측치도 존재x 
```

```python
# df와 customer_master 조인 결합  -> 조건 customer_id
df = pd.merge(
    df, customer_master, 
    on = 'customer_id', 
    how = 'inner'
)
```

```python
# 시계열데이터 payment_date를 컬럼을 타입을 변경 
# 문자를 시계열로 변경 ( datetime.strptime() // to_datetime() )
pd.to_datetime(df['payment_date'])
```

```python
# datetime 라이브러리의 strptime()을 이용하여 데이터 타입 변환 
from datetime import datetime
```

```python
df['payment_date'][0]
```

```python
# datetime 안에 strptime() 함수는 pandas에 to_datetime()과는 다르게 스리즈 형태 변환 불가능
# datetime.strptime(df['payment_date'], '%Y-%m-%d %H:%M:%S')
# 스리즈 데이터에서 value들을 하나씩 뽑아서 함수에 넣는다 map()
df['payment_date'].map(
    # x에 대입이 되는 데이터는 문자
    lambda x : datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
)
```

```python
# 스리즈 데이터에서 문자열 함수를 사용하기 위해 
    # 1. 반복문 사용
    # 2. map() + lambda
    # 문자열 함수를 사용할수 있도록 pandas에서 제공한 Series.str (StringMethods)
# 스리즈에서는 시계열 함수를 사용할 수 있도록 문자열과 같이 DatetimeMethods 변환이 존재 
    # Series.dt --> 해당 데이터셋이 시계열로 시계열 함수 
```

```python
df['payment_date'] = pd.to_datetime(df['payment_date'])
```

```python
df.info()
```

```python
# 요일별 판매가 된 제품들의  총 판매 개수 합계가 어떻게 되는가?
# 요일이라는 새로운 변수를 생성 -> 시계열에서 요일만 따로 추출(strftime(%A|%a|%w))
df['요일'] = df['payment_date'].dt.strftime('%A')
# strftime(%w) -> 0(일), ...
```

```python
df['payment_date'].dt.weekday
# weekday -> 0(월), ...
```

```python
df.head(1)
```

```python
# 요일, item_id를 기준으로 그룹화 -> quantity의 합산 
df[['요일', 'item_id', 'quantity']].groupby(['요일', 'item_id']).sum()
```

```python
# transaction데이터에서 customer_id와 
# customer_master데이터에서 customer_id의 값들이 다른가?

list_1 = transaction['customer_id'].unique()
```

```python
list_2 = customer_master['customer_id'].unique()
```

```python
# 두개의 리스트에서 서로 다른 값을 다른 데이터는 어떻게 되는가?
new_list = []

for customer in list_1:
    if customer not in list_2:
        # 고객의 ID가 list_2에 존재하지 않는다면
        new_list.append(customer)

new_list
```

```python
[ customer for customer in list_1 if customer not in list_2 ]
```

```python
# 집합 데이터 타입 : { } | set() 묶여서 데이터 생성 
# 집합 -> 중복 데이터가 존재하지 않음 
set(transaction['customer_id']) - set(customer_master['customer_id'])
```
