#### class 선언 
- class 키워드를 이용해서 class를 선언 
- class의 이름은 일반적으로 단어의 첫글자를 대문자로 표기

#### class 생성 
- 선언이 된 class를 특정 공간에 저장하는 과정 ( 변수를 생성해서 class을 할당 )
- 생성된 class 안에는 변수와 함수가 존재 ( 같은 변수명을 사용하더라도 각각 따로 저장 )

#### class 내부의 변수, 함수를 호출 
- class가 저장되어 있는 공간(변수명) + ~안에(.) + class에 존재하는 변수(변수명) // 함수(함수명()) 호출

```python
# class 선언
class Class_1():
    # 생성자 함수 (초기화 함수)
    # class 내부에 함수를 선언하는 과정에서 일반적으로 첫번째 매개변수의 이름은 self 로 지정
    # self : 자기 자신(생성되는 class 공간을 의미)
    # __init__ 함수는 class가 생성이 될때 자동적으로 한번만 호출이 되는 특수 함수
    # 따로 호출할 필요가 없음
    # 생성자 함수는 class에서 필요가 없다면 만들지 않아도 무관
    # 생성자 함수를 사용하는 이유 -> class에서 사용할 변수들을 등록하는 과정
    def __init__( self, _a, _b ):
        # class 내부에서 사용할 변수들을 저장
        # self.a -> self(자기자신) + .(~안에) + a(공간의 이름)
        self.a = _a
        self.b = _b
        # return "Hello"
        # __init__ 함수는 class에서 사용할 변수를 저장하는 기능이기때문에
        # return을 만들지 않는다 
        # ( class 생성 시 자동 실행이기 때문에 return으로 받아오는 데이터를 저장 x )
    # def __init__( self, input_a, input_b ):
    #     self.a = input_a
    #     self.b = input_b
    # def __init__( self, a, b ):
    #     self.a = a
    #     self.b = b
    def add_value(self):
        # class 내부에 저장된 self.a와 self.b를 더해서 되돌려준다.
        result = self.a + self.b
        return result
```

```python
# class 생성 -> 선언이 된 클래스를 특정 공간에 저장(할당)
a_class = Class_1(10, 5)
# class 생성 -> __init__ 함수가 자동 호출
# _a = 10 , _b = 5
# self.a = _a(10) , self.b = _b(5)
b_class = Class_1(3, 7)
# class 생성 -> __init__ 함수가 자동 호출
# _a = 3 , _b = 7
# self.a = _a(3) , self.b = _b(7)
```

```python
# class 내부에 존재하는 변수를 호출 
# 공간 안에 변수 호출
print(a_class.a)        # 3(x) 10(o)
print(a_class.b)        # 7(x) 5(o)
print(b_class.a)        # 3
print(b_class.b)        # 7
```

```python
# class 내부에 존재하는 함수를 호출 
print(a_class.add_value())  # 15
print(b_class.add_value())  # 10
```

```python
# 계산기 class 선언 (사칙연산)
class Calculate:
    # 생성자 함수의 선언
    def __init__(self, _a, _b):
        self.a = _a
        self.b = _b
    
    # class에 저장된 변수를 더해서 되돌려주는 함수 
    def add(self):
        result = self.a + self.b
        return result
    
    def sub(self):
        result = self.a - self.b
        return result
    
    def multi(self):
        result = self.a * self.b
        return result
    
    def dev(self):
        result = self.a / self.b
        return result
    # 처음에 입력했던 데이터가 변하는 부분은 하나도 존재하지 않는다. 
```

```python
# class 생성 
cal_1 = Calculate(10, 5)    # a에 10 대입, b에 5 대입
cal_2 = Calculate(3, 7)     # a에 3 대입, b에 7 대입
```

```python
class Test2:
    def __init__(self, _a):
        self.a = _a

    # 원본의 데이터를 수정할 것인가? 아니면 결과를 되돌려줄것인가? 
    # 수정 할지 말지 매개변수를 하나 생성( 두가지의 경우니 bool )
    def func_1(self, inplace = False):
        # inplace 매개변수는 원본의 데이터를 수정할것인가를 지정하는 입력값
        # False 기본값으로 사용하는 이유는 -> 원본 데이터를 수정하는 부분은 함수를 호출하는 사람이 선택하도록
        if inplace:
            # 원본의 데이터를 수정 
            self.a = self.a ** 2
        else:
            # 원본의 데이터는 유지 
            # 연산의 결과물을 되돌려준다 -> ( 확인 // 원본 데이터는 그대로 유지하면서 결과물을 다른 곳에 저장 )
            result = self.a ** 2
            # print(result)
            return result
            # return 과 print()의 차이 -> 
                # print() -> 단순 출력 -> 해당 데이터를 저장이 불가능
                # return -> ipynb 파일에서는 확인 가능 -> 데이터 저장 가능
```

```python
# 인자에 * -> list의 주소를 참조하는가? -> 아니다
# *매개변수 --> 인자들은 하나로 묶어준다. (패킹)
# *인자 --> 묶여 있는 데이터를 풀어준다. (언패킹)
#       --> 리스트의 주소를 참조하는것이 아니라 리스트 내부의 원소의 주소를 참조
#       --> func(*[1,2,3]) --> func(list[0],list[1],list[2])
#       --> a = list[0], b = list[1], c = list[2] (o) 

def test_func(a, b, c):
    a += 1
    b += 2
    c += 3
    return [a, b, c]
# 언패킹 할 데이터셋이 1차원 데이터인 경우
a_list = [1,2,3]

# 함수의 결과를 출력
print(test_func(*a_list))
# 인자를 출력
print(a_list)

# 언패킹으로 입력이 된 원본의 데이터를 수정이 되지 않는다. 
# ( list 자체의 주소를 참조하지 않는다 )
```

```python
# 리스트의 주소를 참조한다 sample
b_list = a_list

b_list.append(1)

print(a_list)
print(b_list)
```

```python
# 함수의 인자값에 언패킹할 데이터가 2차원 리스트인 경우 
def test_func2(a, b, c):
    a.append(1)
    b.pop()
    c.reverse()
    return [a, b, c]

c_list = [
    [1,2], [3,4], [5,6]
]
print( test_func2(*c_list) )
# a = [1,2], b = [3,4], c = [5,6]
print(c_list)
```

#### 상속 
- 클래스를 생성할때 다른 클래스의 기능을 가지고 와서 사용하는 방법 
    - 오버라이드(override)
        - python에서 기본적으로 지원
        - 부모 클래스( 기능을 넘겨주는 클래스 )의 메서드(함수)의 이름을 다른 기능으로 변경하는 방법 
        - 부모 클래스에 a 함수가 있고 자식 클래스에 a 함수를 선언하면 자식 클래스에서는 a 함수가 덮어씌워진다. ( 부모 클래스의 a함수는 그대로 유지 )
    - 오버로딩(overloading)
        - python에서 기본적으로 지원해주지 않음 
        - 같은 메서드(함수)를 여러개 만드는 방법
        - 같은 함수를 호출하는데 인자의 타입이 다른 경우 -> 다른 연산을 하겠다

#### 연습 문제 
1. 클래스 선언 
    - 클래스의 이름은 Bank
    - 생성자 함수 
        - 매개변수 4개 (self 미포함)
            - _name : 기본값x   (통장 주인의 이름) --> self.name 저장
            - _birth : 기본값x  (통장 주인의 생년월일) --> self.birth 저장
            - _cost : 기본값 0  (통장의 잔액)       --> self.cost 저장
            - _log : 기본값 []  (통장의 거래 내역)  --> self.log 저장
    - 함수 생성 
        - 입금 함수 (add_cost)
            - 매개변수(1개) --> 입력값?? -> 변할 수 있는 데이터는 우선적으로 모두 매개변수
                - 금액(_c) : 기본값x (필수 항목이냐 선택 항목)
            - self.cost를 _c만큼 증가 (누적합)
            - log에 dict형태로 {'type' : '입금', 'cost' : _c, 'total_cost' : self.cost} 추가
        - 출금 함수 (sub_cost)
            - 매개변수 (1개)
                - 금액(_c) : 기본값x
            - self.cost보다 _c가 작거나 같은 경우에만 self.cost를 감소 _c가 더 큰 경우에는 self.cost를 유지하면서 '잔액이 부족합니다'를 출력 
            - self.log에 출금이 성공했을때만 {'type':'출금', 'cost':_c, 'total_cost' : self.cost} 추가

```python
class Bank():

    # class 변수를 생성 ( 같은 class를 생성하는 객체들끼리 서로 공유할 수 있는 변수 )
    total_cost = 0      # 모든 유저가 입금 / 출금시 total_cost의 변화 
    #                   ( class 생성 시 _cost를 total_cost에 누적합)
    user_cnt = 0        # class 생성 시 1씩 증가
    # user_list = []
    # 생성자 함수 : 매개변수 4개 
    # def __init__(self, _name, _birth, _cost = 0, _log = []):
    def __init__(self, _name, _birth, _cost = 0):
        self.name = _name
        self.birth = _birth
        self.cost = _cost
        # self.log = _log
        self.log = []
        # class 변수에 접근하는 방법
        Bank.total_cost += _cost
        Bank.user_cnt += 1
        # self.user_list = Bank.user_list --> 쓰레기 코드 (필요가 없는 코드)
    # 입금 함수 
    def add_cost(self, _c):
        # 단순하게 self.cost를 _c만큼 증가 
        self.cost += _c
        # self.cost = self.cost + _c
        # 로그 내역 추가 
        dict_data = {
            'type' : '입금', 
            'cost' : _c, 
            'total_cost' : self.cost
        }
        # class 변수 은행의 총 잔액을 증가
        Bank.total_cost += _c
        self.log.append(dict_data)
        print(f"입금이 완료되었습니다. 현재의 잔액은 {self.cost}입니다")
    # 출금 함수 
    def sub_cost(self, _c):
        # 현재의 잔액보다 _c가 작거나 같아야 출금이 완료
        if self.cost >= _c:
            # 출금이 가능한 상황 
            self.cost -= _c
            dict_data = {
                'type' : '출금',
                'cost' : _c, 
                'total_cost' : self.cost
            }
            # 거래내역을 추가 
            self.log.append(dict_data)
            # class 변수 은행 총 잔액을 감소
            Bank.total_cost -= _c
            print(f"출금이 완료되었습니다. 현재의 잔액은 {self.cost}입니다")
            # 출금이 정상적으로 작동했을때 1을 되돌려준다. 
            return 1
        else:
            # 출금이 불가능한 상황
            # self.cost 유지 --> 아무 행동도 하지 않는다. 
            print('잔액이 부족합니다.')
            # 출금 실패 시 0을 되돌려준다. 
            return 0
    # user_info() 함수를 생성하여 이름과 생년월일 현재 잔액을 출력하는 함수
    def user_info(self):
        print(f"이름 : {self.name}, 생년월일 : {self.birth}, 잔액 : {self.cost}")
```

```python
# 상속 / 오버라이드 

# 부모 클래스 선언 
class A:
    def __init__(self, _name, _age):
        self.name = _name
        self.age = _age
        self.a = 10
    def user_info(self):
        print(f"유저의 이름은 {self.name}이고 나이는 {self.age}입니다")
        print(self.a)
```

```python
user_1 = A('kim', 30)
```

```python
user_1.user_info()
```

```python
# 자식 클래스를 하나 생성 -> A클래스의 기능을 상속 받는다
class B(A):
    # B 클래스는 A 클래스의 함수들을 사용이 가능
    def __init__(self, _name, _age, _addr):
        # _name, _age는 부모 클래스에서 변수로 저장하는 기능 존재 
        # self.name = _name
        # self.age = _age
        # self.a = 20
        # 부모클래스(super()) 안에(.) 생성자 함수(__init__)를 호출
        super().__init__(_name, _age)
        self.addr = _addr
    def user(self):
        print(f"이름 : {self.name}, 나이 : {self.age}, 주소 : {self.addr}")

# 자식 클래스에서 부모 클래스를 상속 받는다 -> 기능(생성자 함수 제외)을 물려받는다
# 생성자 함수의 특징때문에 발생하는 일 -> 해당 이름의 클래스가 생성이 될때만 실행
# 부모 클래스에서 사용이되는 변수들을 모두 사용을 하기 위해서는 
    # 부모 클래스와 같이 자식 클래스에서 변수를 모두 생성 
    # 부모 클래스의 생성자 함수를 호출 
```

```python
user_2 = B('park', 20, '서울')
```

```python
# user_2에는 함수가 user()사용 가능 , user_info() 함수도 사용이 가능
user_2.user()
```

    - 해당하는 일을 얼마나 하는가? -> 매개변수 
    - 일의 종류에 따라 시급이 지정이 되어있고 시급 * 시간 -> Bank에 add_cost()함수를 이용해서 통장의 잔액이 증가 
- 돈을 쓴다 함수 
    - 어떤 물건을 구매할 것인가? -> 매개변수 
    - 구매하는 물건에 따라 Bank에 sub_cost() 함수를 이용해서 잔액을 감소 
        - Bank의 sub_cost() 함수에서 성공하는 경우와 실패하는 경우에 return 데이터를 0과 1로 되돌려준다 수정 
    - sub_cost() 함수의 결과 값에 따라서 물건을 구매 성공/ 실패 여부를 출력 
- 오버라이드 기능 사용 
    - Bank 클래스에서 user_info() 함수를 User에서 같은 메서드를 생성하여 덮어씌워준다.
        - 이름과 나이, 현재 잔액과 구매한 물품 목록을 보여준다. 
            - 구매한 물품의 목록은 User 클래스에서 생성자 함수에서 변수를 생성

```python
from typing import override

# 일의 종류 -> 
    # 함수 안에서 여러 조건식을 이용하여 일의 종류, 시급들을 지정 
    # DataBase(데이터들을 모아두는 공간)을 이용하여 데이터를 로드 
    # key : value 의 형태를 이용하여 일의 종류 : 시급 이라는 형태의 데이터를 사용
# 구매 할 수 물품 리스트 ->
    # dict형태의 key : value의 형태를 이용하여 지정이 가능 
    # 함수 안에서 여러 조건식 사용 가능
    # DB 안에 데이터를 이용하여 사용 가능 

# 두개의 데이터셋은 dict 형태로 class 변수 로 지정
# ( 같은 class를 생성했을때 일의 종류와 물건의 종류는 서로 공유가 가능하게 위해 )

class User(Bank):

    # 일의 종류와 시급 dict 데이터를 class 변수로 생성 
    work_type ={
        'A' : 11000, 
        'B' : 15000, 
        'C' : 20000
    }
    # 구매하고 싶은 물건
    item_list = {
        '텀블러' : 50000, 
        '스위치' : 730000, 
        '태블릿 거치대' : 20000 , 
        '헤드셋' : 520000, 
        '노트북' : 2000000
    }

    # 생성자 함수 
    def __init__(self, _name, _birth, _cost):
        # _name, _birth, _cost는 부모 클래스의 생성자 함수를 호출해서 인자로 사용
        super().__init__(_name, _birth, _cost)
        # super().__init__() -> 생성이 되는 변수의 개수는? 4개 이름, 생년월일, 잔액, 거래내역
        # 구매한 아이템의 목록 -> 빈 리스트로 생성 
        self.items = []

    # 일을한다... 함수 (매개변수 : 일의 종류, 시간)
    def work(self, _type, _time):
        # _type이 wotk_type에 데이터가 존재하는가?
        # dict 데이터에서 key 값들 중에 _type의 값이 포함되어있는가?
        # in 비교 연산자 -> 단일 데이터 in 1차원 데이터 -> 1차원 데이터 원소 중 단일 데이터가 포함되어있는가?
        if _type in User.work_type:
            # 일의 종류 데이터에 _type이 포함되어 있는 경우
            # 시급 데이터를 로드하고 _time과 * 연산을 한다.
            # 시급 -> work_type 데이터가 존재 
                # { 'A' : 11000, 'B' : 15000, 'C' : 20000 } <- User.work_type
            # _type이 'A' 라고 입력값이 들어왔다면 
                # User.work_type['A'] -> 11000 --> A라는 일의 시급 
                # _time 매개변수가 의미하는것은 몇시간 일했는가?
                # 시급 * 시간 --> 일당
                # 일당을 통장에 입금
            amount = User.work_type[_type] * _time  # 일당을 계산
            # print(User.work_type)
            # print(User.work_type[_type])  # 시급 
            # print(amount)       # 일당을 출력
            # 일당이 계산이 되었으니 일당을 통장
            # (Bank-> 부모클래스 -> super())에(~안에) 입금(add_cost( 금액 ))
            super().add_cost( amount )   # -> add_cost라는 함수는 입금완료 출력이 존재
        else:
            # 일의 타입이 맞지 않는 경우 
            # print('일의 타입이 맞지 않습니다.')
            # 잘못된 데이터가 들어왔을때 에러 발생 raise
            # TypeError (데이터의 타입이 맞지 않는다.)
            # ValueError (데이터의 값이 잘못되었다.)
            raise ValueError( "저장되어 있는 일의 종류가 아닙니다." )
    # 물건을 구매하자 (스트레스 해소) -> 매개변수 1개 (물건의 종류)
    def buy_item(self, _item):
        # work() 함수에서는 일의 종류가 포함되어 있는가? -> in이라는 비교연산자 사용 

        # try ~ except : 코드를 실행하는데 예외가 발생했을때 특정한 행동을 할때 사용
        try :
            # 시도할 코드 
            # 물건의 가격을 로드 
            # 물건의 가격은 User 클래스 안에 item_list에 dict 형태로 존재 
            # -> key을 이용해서 가격을 추출 dict[key]
            # key로 사용될 데이터는 ? -> _item
            # print(User.item_list)
            # print(User.item_list[_item])    # 물건의 가격을 추출
            # 가격을 알았으니 출금 (Bank(super()) 안에 sub_cost( 금액 ))
            # 잔액이 부족하면 0을 되돌려주고 정상적으로 출금이 되면 1 되돌려준다.
            sub_res = super().sub_cost( User.item_list[_item] )
            # sub_res가 1이면 구매 성공 -> items에 구매한 물건을 추가 
            # if sub_res == 1:
            if sub_res:
                # 구매 성공 
                self.items.append(_item)
                print('구매 성공')
            else:
                # 구매 목록에 추가하지 않는다. -> 아무 행동도 하지 않는다
                print('구매 실패')
        except Exception as e:
            # Exception : 에러의 메시지
            # as : 별칭을 짓겠다
            # e : 별칭 명
            # 에러의 종류를 출력
            print(type(e).__name__)
            print(e)   # 어떤 에러가 발생했는지 이유를 print() 출력
            # try 영역에서 실행되는 코드가 예외가 발생했을때만 실행이 되는 영역
            print('물건의 정보가 없습니다.')
    # 오버라이드 : 부모클래스의 기능의 이름을 덮어씌워서 사용 
    # (부모 클래스에서는 1이라는 기능을 하고 자식 클래스에서는 2이라는 기능을 한다. ) -> 
    # 부모 클래스의 함수의 기능은 그대로 유지 
    # 이름, 생년월일, 잔액, 구매한 물건의 목록을 출력 
    @override     # 문법적으로 오버라이드 함수입니다. -> 가독성이 좋게 표시 ()
    def user_info(self):
        print( f"""이름 : {self.name}, 생년월일 : {self.birth}, 
            잔액 : {self.cost}, 구매한 물건의 목록 : {self.items}""")
    # User 클래스에서는 이름 생년월일 잔액 물건의 목록을 출력
    # Bank 클래스에서는 이름 생년월일 잔액을 출력
```

```python
# 오버라이드는 일반적으로 자식 클래스에서 부모 클래스의 함수명을 그대로 사용
# python 3.12 버전 이하 
# from typing_extensions import override
# 3.12 버전 이후 
# from typing import override
```

```python
# 오버로딩 -> 일반적으로 지원하지 않는다. -> 매개변수가 인자를 가변으로 받는다.
from functools import singledispatch

@singledispatch
def show(data):
    print(f"기본형 함수 : {data}")

# show() 함수의 data 매개변수에 인자의 타입이 만약에 int 라면 
@show.register(int)
def _(data):
    print(f"정수형 함수 : {data}")

@show.register(list)
def _(data):
    print(f"리스트형 함수 : {data}")
```

```python
show('a')
show(10)
show([1,2,3,4])
```

```python
def show2(data):
    if type(data) == str:
        print(f"기본형 함수 : {data}")
    elif type(data) == int:
        print(f"정수형 함수 {data}")
    elif type(data) == list:
        print(f"리스트형 함수 {data}")
```

```python
show2('a')
show2(10)
show2([1,2,3,4])
```
