### 데이터베이스와 python의 연결
1. 데이터베이스와 연결을 하기 위해서 라이브러리를 설치해야한다.
2. 라이브러리를 로드한다.
3. 데이터베이스 서버와의 연결을 해준다.
    - 서버의 정보
        - 서버의 주소
        - 서버의 포트(3306)
        - 아이디 (root)
        - 비밀번호 (비밀번호가 없으면 생략 가능)
        - 사용할 데이터베이스명 (multicam)
4. cursor(가상공간)를 생성
5. SQL query문을 작성하여 질의를 DB서버에 보낸다. 
    - select(조회)
        - return 데이터가 존재
        - fetchall() 함수를 이용해서 결과 값을 돌려받는다.
    - insert, update, delete 
        - return 데이터가 존재하지 않는다.
        - 실제 데이터가 변화하는 부분이기 때문에 Cursor가 변환이 된다. -> DB 서버와 Cursor의 동기화가 필요하다.
6. 서버와의 연결을 종료한다.

```python
import pymysql
```

```python
# DB서버연동
# 외부 DB에 접속시 IP 필요
# 내 컴퓨터 db server 주소 -> 127.0.01 | localhost

_db = pymysql.connect(
    host='127.0.0.1', #서버주소
    port=3306, #서버포트(숫자로입력)
    user='root', # DB Server 접속 아이디
    password='1234', # 비밀번호 (비밀번호 없는 경우 생략 가능)
    db='multicam' # 데이터베이스명

)
```

```python
# 가상공간인 Cursor 생성

# 기본값으로 커서 생성 -> select 문으로 데이터를 로드할때 결과 데이터의 타입이 tuple
cursor = _db.cursor()

# DictCursor 옵션을 사용해서 커서를 생성
cursor2=_db.cursor(pymysql.cursors.DictCursor)
```

```python
# query문을 작성 -> 문자열 데이터로 작성
select_query="""
    select 
    * 
    from 
    `emp`
"""
# select_query를 cursor에게 보낸다 (질의)
cursor.execute(select_query)
cursor2.execute(select_query)

# execute() 함수의 리턴은 select인 경우 인덱스의 수 출력
```

```python
# cursor안에 select의 결과가 저장되어 있어서 해당 데이터를 가져온다. (fetchall())
# fetchall() 함수를 호출 시 cursor의 데이터를 가져오고 cursor에 데이터는 제거
select1=cursor.fetchall()
select2=cursor2.fetchall()
```
