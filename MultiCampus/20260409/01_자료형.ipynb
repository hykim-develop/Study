```python
# 반복문을 이용
# for(자료형 데이터에서 각각의 원소들을 추출하여 변수에 대입하여 반복)
# while(특정 조건식이 거짓이 될때까지 반복)
for i in sample_data:
    # i에 들어오는 데이터는?? -> 아 dict의 형태의 데이터를 for문에서 사용하면 key가 대입
    # print(i)
    # print(sample_data[i])
    print(sample_data[i][1])
```

```python
# sample_data에서 key 값들의 목록 -> keys()
keys = list(sample_data.keys())
# while문의 초기값
i=0

#while문 생성
#while i < 3 : # 3이라는 의미는? saple_data의 길이
while i < len (sample_data): 
    # 더 좋은 방법 : 동적 할당 -> sample_data의 길이가 변하더라도 사용 가능
    # i의 값의 의미는? -> key들의 인덱스
    # print(keys[i])
    print(sample_data[keys[i]])
    print(sample_data[keys[i]][1])

    # i 값을 1씩 증가
    i+=1
```

```python
# list에서 sort() 함수 -> 리스트의 각각의 원소들을 비교하여 가장 작은 데이터를 처음으로 가장 큰 데이터를 마지막에 배치하여 오름 차순 정렬하는 함수

sample_list=[30,10,50,7]

# 가장 작은 숫자부터 대입이 될 빈 리스트가 생성

sort_list = []

# sample_list의 길이만큼 반복 실행하는 반복문을 생성

for i in range(len(sample_list)):
    # 기준이 되는 list의 첫번째 원소를 변수에 저장 (첫번째 데이터가 가장 작은 데이터라 생각하고 대입)
    min_value=sample_list[0]
    for j in range(1, len(sample_list),1):
    # 2번째 데이터부터 min_value와 비교
    # min_value -> 30과 sample_list[1] -> 10를 비교 --> min_value는 10 대입
    # min_value -> 10과 sample_list[2] -> 50를 비교 --> min_value는 유지 (데이터의 변경이 없다.)
    # min_value -> 10과 sample_list[3] -> 7을 비교 --> min_value에 7 대입
    # min_value와 크기 비교
        if min_value > sample_list[j]: 
            # min_value를 sample_list[j]으로 수정
            min_value=sample_list[j]
    # sort_list에 min_value를 추가 
    # print(min_value)
    # break
    sort_list.append(min_value)
    # sample_list에서 7을 제거 (remove())
    sample_list.remove(min_value)
    print(sample_list)
    print(sort_list)
    # j는 sample_list의 인덱스 (두번째 데이터)
```

```python
sample_list=[30,10,50,7]
sort_list=[]

# min() : 여러개의 데이터중 가장 작은 값을 반환
# sample_list에서 가장 작은 값을 추출 -> sort_list에 추가 -> sample_list에서 제거
# sample_list의 길이만큼 반복 실행
for i in range(len(sample_list)):
    # min() 함수를 이용하여 sample_list에서 가장 작은 데이터를 추출하여 변수에 저장
    min_value=min(sample_list)
    # print(min_value)
    # break
    # sort_list에 추가
    sort_list.append(min_value)
    # sample_list에서 제거
    sample_list.remove(min_value)

print(sort_list)
```

```python
# 내림차순 정렬 -> 리스트에서 가장 큰 데이터를 추출
# max() : 여러개의 데이터에서 가장 큰 값을 반환

sample_list=[30,10,50,7]
sort_list=[]

for i in range(len(sample_list)):
    max_value=max(sample_list)
    sort_list.append(max_value)
    sample_list.remove(max_value)
print(sort_list)
```
