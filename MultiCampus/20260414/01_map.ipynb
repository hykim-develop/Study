```python
df2 = df.copy()   # 원본데이터 백업 
```

```python
# 단일 데이터에서는 type이 문자이기때문에 strip() 사용 가능
# iloc[인덱스의 위치, 컬럼의 위치]
# df.iloc[2, 0]   # index의 위치만 반복실행하면서 1씩 증가시켜 strip()
# 5번 반복 -> 데이터프레임의 row의 개수 -> 길이를 구하는 함수(len())
# -> len(데이터프레임) : 행의 개수를 출력 
for i in range(len(df2)):
    # i는 0부터 4까지 1씩 증가시키면서 대입 (데이터프레임의 인덱스)
    # print(df.iloc[i, 0])
    # print(df2.iloc[i, 0].strip())
    # df2.iloc[i, 0]데이터를 좌우의 공백을 제거하고 
    # df2.iloc[i, 0]에 대입 
    df2.iloc[i, 0] = df2.iloc[i, 0].strip()

df2
```

```python
df2['product'].value_counts()
```

```python
# while문으로 같은 작업
df3 = df.copy()
```

```python
# 초기값 설정 (인덱스)
i = 0

while i < len(df3):
    print(df3.iloc[i, 0].strip())

    # 인덱스(i)를 1씩 증가
    i += 1
```

```python
i = 0

# 무한 반복
while True:
    # df3.iloc[i, 0]
    # 반복 실행 할때마다 i의 값은 1씩 증가 
    # 인덱스의 범위가 벗어나면 에러가 발생 -> 예외 처리를 이용해서 반복문을 종료
    # try ~ except문
    try:
        print(df3.iloc[i, 0].strip())
        i += 1
    except:
        # 위의 코드가 에러가 발생한다면 반복문을 강제 종료
        break
```

```python
# map() : 1차원 데이터에서 각각의 원소들을 함수에 입력하여 데이터를 수정&연산 한다. 
# map() 함수는 python 기본 제공이 있고 Series 내장함수로도 존재
# Series.map( 데이터의 형태를 변화시킨 함수 )
def data_strip(x):
    return x.strip()

df3['product'].map(
    # 각각의 원소들을 입력 받아서 strip() 함수를 적용
    data_strip
)           # 결과는 Series
```

```python
# map() + lambda 함수 
df3['product'].map(
    lambda x : x.strip()
).value_counts()
```

```python
# Series.strip() --> 에러 발생 (Series에는 strip()이 존재하지 않는다)
# 문자열의 객체 함수에 접근하는 방법 Series.str.문자열함수
# type의 변화 -> 데이터프레임(df3) -> 스리즈(df3['product'])
#               -> 문자열메서드(df3['product'].str)
#               -> 스리즈(df3['product'].str.strip())
df3['product'].str.strip()
```

```python
series_1 = pd.Series(
    ['cat', 'dog', np.nan]
)
```

```python
series_1
```

```python
# Series의 데이터들을 모두 대문자로 변환 
series_1.str.upper()
```

```python
# map() 함수를 이용하여 문자를 대문자로 변환 
# 결측치의 타입 -> float -> upper() 함수가 존재하지 않는다(에러 발생)
# map() 함수에 na_action 매개변수 -> 결측치 데이터에 함수를 적용할것인가
series_1.map(
    lambda x : x.upper(), na_action = 'ignore'
)
```

```python
# 일반적으로 요일 데이터를 숫자형으로 표시 
# 0 -> 일, 1 -> 월, ... 6 -> 토
week_series = pd.Series([0, 1, 2, 3, 4, 5, 6])
week_series
```

```python
# week_series에 데이터들을 요일(월, 화, 수, ... ,일)로 변경
# 0을 2진수로 표현하면? -> 0b0
# 6을 2진수로 표현한다면 -> 0b110
# 문자와 숫자의 메모리에 저장이 되는 크기를 비교하면 숫자가 훨씬 작다. 
def week_change(x):
    # x에는 0부터 6까지의 숫자들이 대입 
    if x == 0:
        result = '일'
    elif x == 1:
        result = '월'
    elif x == 2:
        result = '화'
    elif x == 3:
        result = '수'
    elif x == 4:
        result = '목'
    elif x == 5:
        result = '금'
    else:
        result = '토'
    return result

week_series.map(
    week_change
)
```

```python
# 요일은 0~6 숫자로 이루어져있다? -> 이 값들을 인덱스로 생각하면? 
# ['일', '월', ..., '토']
def week_change2(x):
    week_list = ['일', '월', '화', '수', '목', '금', '토']
    return week_list[x]

week_series.map(
    week_change2
)
```

```python
# 딕셔너리 데이터를 map() 함수에서 활용 
week_dict ={
    0 : '일', 
    1 : '월', 
    2 : '화', 
    3 : '수', 
    4 : '목', 
    5 : '금', 
    6 : '토'
}
```

```python
week_series.map(
    week_dict
)
```

```python
week_series.replace(
    week_dict
)
```

```python
gender_series = pd.Series(
    ['f', 'm', '-']
)
```

```python
# f를 female로 변경하고 m은 male 변경 
# if문 한줄 표현  ->   참인경우 출력데이터 if 조건식 else 거짓인경우 출력데이터

gender_series.map(
    lambda x : "female" if x == 'f' else 'male' if x == 'm' else np.nan
)
```

```python
list_1 = [1,2,3,4,5,6]
# 짝수만 추출하여 새로운 리스트

# list_1의 각각의 원소들을 x에 대입하여 2로 나눈 나머지가 0이라면 리스트에 추가
new_list = [ x for x in list_1 if x % 2 == 0 ]

print(new_list)
```

```python
new_list2 = []

for i in list_1 :
    if i % 2 == 0:
        new_list2.append(i)

new_list2
```

```python
# [ 변수명 for (조건식) ] -> 조건식이 참인 경우의 데이터를 리스트에 대입 
# [ (조건식) for ] -> 모든 리스트의 데이터를 활용하고 조건식에 맞춰서 대입 ()
# 홀수의 데이터는 홀수  // 짝수의 데이터는 짝수
new_list3 = [ '홀수' if x % 2 == 1 else '짝수' for x in list_1   ]
print(new_list3)
```

```python
new_list4 = ['홀수' if x % 2 == 1 else '짝수' for x in list_1 if x > 3]
print(new_list4)
```
